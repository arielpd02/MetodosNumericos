```python
import numpy as np
import sys
import matplotlib.pyplot as plt
import timeit

x = np.linspace(-10,10,100)
y = (1-np.cos(x))/x**2
plt.plot(x,y, label='$(1-cos(x))/x^2$')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
```
